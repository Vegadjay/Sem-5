# Step 1: Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("OnlineRetail.csv", encoding='ISO-8859-1')
data.head()
```

```python
data
```

# Step 2: Drop Missing Invoices or Descriptions

```python
data.dropna(subset=['InvoiceNo', 'Description'])
data.shape
```

```python
data
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
data = data[~data['InvoiceNo'].astype(str).str.startswith('C')]
data.shape
```

```python
data
```

# Step 4: Keep Only Positive Quantities

```python
data = data[data['Quantity'] > 0]
data.shape
```

```python
data
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
data['Description'] = data['Description'].str.lower().str.strip()
data[['Description']].head()
```

```python
data
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
data = data[data['Country'] == 'United Kingdom']
data['Country'].value_counts()
```

```python
data
```

# Step 7: Remove Duplicates

```python
data = data.drop_duplicates()
data.shape
```

```python
data
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']
data[['Quantity', 'UnitPrice', 'TotalPrice']].head()
```

```python
data
```

# Step 9 : Check for Top 10 Selling Products

```python
top_products = (data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10))

top_products
```

# Step 10 : Identify Most Active Customers

```python
most_active_customers = (data.groupby('CustomerID')['InvoiceNo'].nunique().sort_values(ascending=False).head(10))

most_active_customers
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
import matplotlib.pyplot as plt

data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

data['Month'] = data['InvoiceDate'].dt.month
data['Hour'] = data['InvoiceDate'].dt.hour

month_counts = data['Month'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
month_counts.plot(kind='bar', color='skyblue')
plt.title('Transactions per Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.show()

hour_counts = data['Hour'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
hour_counts.plot(kind='bar', color='lightgreen')
plt.title('Transactions per Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.show()
```

# Step 12: Convert to Basket Format 

Example
<table border="1">
    <thead>
        <tr>
            <th>Transaction</th>
            <th>Milk</th>
            <th>Bread</th>
            <th>Butter</th>
        </tr>
    </thead>
    <tbody>
        <tr><td>T1</td><td>1</td><td>1</td><td>1</td></tr>
        <tr><td>T2</td><td>0</td><td>1</td><td>1</td></tr>
        <tr><td>T3</td><td>1</td><td>0</td><td>0</td></tr>
        <tr><td>T4</td><td>1</td><td>1</td><td>0</td></tr>
        <tr><td>T5</td><td>0</td><td>1</td><td>1</td></tr>
    </tbody>
</table>


```python
basket = (data
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum()
          .unstack()
          .fillna(0))

basket = basket.applymap(lambda x: 1 if x > 0 else 0)

basket.head()
```

# Step:13 : Save Basket Format into CSV file

```python
basket.to_csv('basket_format.csv', index=True)
print("Basket format saved as 'basket_format.csv'")
```
