### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
x = ["PassengerId", "Survived", "Pclass", "Age", "SibSp", "Fare"]

for i in x:
    print('\t')
    print(i)
    print(f'Mean is {data[i].mean()}')
    print(f'Mode is {data[i].mode()[0]}')
    print(f'Max is {data[i].max()}')
    print(f'Min is {data[i].min()}')
```
